# 2. Divisão do conjunto de dados

Nesta fase, farei a divisão do conjunto de dados em duas partes: treinamento e teste. A optei por alocar 80% para treinamento e 20% para teste.

# 3. Seleção de algoritmos de classificação

Aqui está uma lista de algoritmos de classificação acompanhada de uma breve explicação, que serão submetidos a treinamento e avaliação na próxima etapa.

`Naive Bayes`: assume que as características de um exemplo são independentes umas das outras. Isso significa que a probabilidade de um exemplo pertencer a uma classe não é afetada pelas características de outros exemplos. Para classificar um exemplo, o algoritmo calcula a probabilidade de cada classe, dado um conjunto de características. A probabilidade de uma classe é calculada como o produto das probabilidades das características dessa classe.
Ou seja, o algoritmo olha para as características de um exemplo e calcula a probabilidade de cada classe. A classe com a maior probabilidade é a classe que o exemplo é mais provável de pertencer.

`k-Nearest Neighbors`: classifica um exemplo com base nos seus vizinhos mais próximos. Ele calcula a distância entre o exemplo e todos os outros exemplos no conjunto de dados e atribui o exemplo à classe mais comum entre seus vizinhos mais próximos. Ou seja, o algoritmo olha para os exemplos mais próximos de um exemplo e vê a classe mais comum desses exemplos. A classe mais comum é a classe que o exemplo é mais provável de pertencer.

`Support Vector Machine (SVM)`: encontra uma fronteira que separa os exemplos de duas classes. Ele tenta maximizar a distância entre a fronteira e os exemplos de cada classe. Ou seja, o algoritmo encontra uma linha (ou plano) que separa os exemplos de duas classes. Os exemplos de cada classe estão localizados em lados opostos da fronteira.

`Decision Trees`: cria uma árvore de decisão para classificar exemplos. Cada nó da árvore representa uma característica e cada ramo representa um valor possível para essa característica. O exemplo é classificado seguindo o caminho da árvore até uma folha. Ou seja, o algoritmo cria uma árvore com perguntas sobre as características dos exemplos. A resposta a cada pergunta leva a uma nova pergunta ou a uma classificação.

`Random Forest` é um conjunto de árvores de decisão. Ele combina as previsões de todas as árvores para obter uma classificação final.
Ou seja, o algoritmo cria várias árvores de decisão e combina as previsões dessas árvores para obter uma classificação.

`Multilayer Perceptron`: é uma rede neural que aprende a relacionar as características de um exemplo com sua classe. A rede é treinada com um conjunto de dados de exemplos classificados. A rede aprende a relacionar as características dos exemplos com a classe dos exemplos.

`Gradient Boosting`: é uma técnica de aprendizado de máquina que constrói modelos de predição em forma de árvore de decisão, de forma sequencial, onde cada árvore corrige os erros do modelo anterior. Visa melhorar o desempenho preditivo, combinando diversas árvores fracas para formar um modelo robusto e preciso.


# 4. Treinamento e avaliação

Nesta fase, irei analisar as métricas de desempenho, incluindo acurácia, precisão, revocação e escore F1 para cada algoritmo. Para uma comparação justa, optarei por não ajustar nenhum parâmetro em nenhum dos algoritmos, garantindo que os resultados reflitam os valores padrões dos parâmetros de cada classificador.

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Instancia o classificador
naive_bayes_model = GaussianNB()

# Treina o modelo com os dados de treinamento
naive_bayes_model.fit(X_train, y_train)

# 3. Faz previsões no conjunto de teste
y_pred = naive_bayes_model.predict(X_test)

# 4. calcula as métricas de desempenho
accuracyNB = accuracy_score(y_test, y_pred)
precisionNB = precision_score(y_test, y_pred)
recallNB = recall_score(y_test, y_pred)
f1NB = f1_score(y_test, y_pred)

# Printa as métricas de desempenho
print(f"Acurácia: {round(accuracyNB*100, 2)}%")
print(f"Precisão: {round(precisionNB*100, 2)}%")
print(f"Revocação: {round(recallNB*100, 2)}%")
print(f"Escore F1: {round(f1NB*100, 2)}%")
```

## Testes com validação cruzada e ajuste de hiperparâmetros


Na fase de seleção do método de validação cruzada, surgiu a dúvida entre utilizar o cross-validation e o grid search. No entanto, ao analisar mais detalhadamente, observei que o grid search não apenas realiza a busca por hiperparâmetros ideais, mas também incorpora a validação cruzada. Dessa forma, optei pelo grid search, pois ele oferece uma abordagem mais abrangente.

Assim, meu foco será aprimorar os hiperparâmetros visando as métrica de "acurácia", "precision e "recall". Tive dificuldades em analisar a métrica de "f1-score", por esse motivo, nessa etapa utilizarei somente as outras. ### Teste `Naive Bayes`


```python

from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import GridSearchCV
import warnings

warnings.filterwarnings(action='ignore')

# Instancia o classificador
gnb = GaussianNB()

# Define a grade de hiperparâmetros
param_grid = {
    'priors': [[0.5, 0.5], [0.6, 0.4], [0.7, 0.3], [0.8, 0.2], [0.9, 0.1]],
    'var_smoothing': [1e-09, 1e-08, 1e-07, 1e-06, 1e-05],
}
# dicionário que guardará as métricas e os scores associados
scores_NB = {}

# Realiza busca em grade com validação cruzada de 10 dobras
for metric in ['accuracy', 'precision', 'recall']:
    grid_search = GridSearchCV(gnb, param_grid=param_grid, cv=10, scoring=metric)
    grid_search.fit(X_train, y_train)

    # Calcula os melhores parâmetros
    melhores_parâmetros = grid_search.best_params_

    # Imprime o resultado
    print(f"Métrica: {metric}")
    print(f"Melhores parâmetros: {melhores_parâmetros}")

    # Instancia o classificador com os melhores parâmetros
    gnb = GaussianNB(priors=melhores_parâmetros['priors'], var_smoothing=melhores_parâmetros['var_smoothing'])

    # Treina o modelo
    gnb.fit(X_train, y_train)

    # Realiza previsões
    y_pred = gnb.predict(X_test)

    # Calcula a métrica de desempenho
    score = accuracy_score(y_test, y_pred)

    # Adiciona o score ao dicionário
    scores_NB[metric] = score

    # Imprime o resultado
    print(f"Score: {round(score*100, 2)}%")
    print(f"-----------------------------------")
```
